```python
#Code to pull data based on date filter(Start & end) as a condition

import requests
from requests.auth import HTTPBasicAuth
import datetime
token = '***********'
ID = '***********'

query = {
    "query": {
        "operator": "AND",
        "value": [
          {
            "field": "signed_up_at",
            "operator": ">",
            "value": datetime.datetime(2021,5,8).timestamp()
          }, 
          {
            "field": "signed_up_at",
            "operator": "<",
            "value": datetime.datetime(2021,5,10).timestamp()
          }
        ]
}}


data = requests.post("https://api.intercom.io/contacts/search", auth = HTTPBasicAuth(token, ID), json=query)
data.json()

```
